# 2. Analisis de Transacciones

## 2.1 Resumen

```python
print(f'There were a total of {data.shape[0]:,} transactions, involving items ranging from'
      + f' {data.shape[1]} to {data.notna().sum(axis=1).min()}.')
```

## 2.2 Tipos de productos

```python
all_items = data.melt()['value'].dropna()
print(f'There were {all_items.nunique()} different products:\n')
print(sorted(all_items.unique()))
```

## 2.3 Productos más comprados

Asume que se compró una sola unidad de cada producto en cada transaccion.

```python
ax = all_items.value_counts().head(15).plot(kind='bar')
ax.set_title('Items comprados con más frecuencia', size=20)
ax.set_ylabel('Cantidad')
```

## 2.4 Productos menos comprados

Asume que se compró una sola unidad de cada producto en cada transaccion.

```python
ax = all_items.value_counts().nsmallest(143).plot(kind='bar')
ax.set_title('Items menos vendidos', size=20)
ax.set_ylabel('Cantidad')
```

## 2.5 Tamaño de la compra
Una "compra" corresponde al conjunto de items únicos comprados en una sola transacción.

```python
basket_sizes = data.notna().apply(sum, axis=1)

ax = basket_sizes.value_counts().plot.bar()
ax.set_title('Distribucion de Tamaños de Compras', size=20)
ax.set_ylabel('Cantidad')
ax.set_xlabel('Numero de items en un sola transaccion')
```

## 2.4 Analizar transacciones más largas


```python
basket_sizes.nlargest(10)
```

```python
largest_transactions = data[basket_sizes > 7]   # se puede modificar el 7
items_in_largest_transactions = largest_transactions.melt()['value'].dropna()

ax = items_in_largest_transactions.value_counts().head(20).plot.bar()
ax.set_title('Apariciones en las 8 transacciones más largas', size=20)
ax.set_ylabel('Numero de transacciones (de 8)')
```

```python
pie_data = items_in_largest_transactions.value_counts()
print(len(pie_data))
ax = pie_data.plot.pie(cmap='autumn', explode=[0.2] * len(pie_data), figsize=(12, 12),
                       autopct=lambda pct: f' {pct * 0.01 * pie_data.sum():.0f} of 8'
)
ax.set_title('Apariciones en las 8 transacciones más largas', size=20, pad=25)
ax.set_ylabel('')
ax.figure.tight_layout()
```

# 3. Analisis de Asociaciones (Association Rule Mining)

## 3.1 Herramienta

Se usa el paquete Python [apyori][2] para implementar el algoritmo [*Apriori algorithm*][3].

[1]: https://en.wikipedia.org/wiki/Association_rule_learning
[2]: https://pypi.org/project/apyori/
[3]: https://en.wikipedia.org/wiki/Apriori_algorithm

## 3.2 Preprocesamiento

El método `apyori.apriori` espera una lista de transacciones (compras). Para encontrar asociaciones/relaciones entre items, las transacciones deben incluir más de 1 item.

```python
baskets = [set(row.dropna()) for _, row in data.iterrows() if row.dropna().size > 1]
baskets[:5]
```

## 3.3 Reglas de Asociación

```python
import apyori

# apyori.apriori devuelve un GENERATOR

association_rules = apyori.apriori(baskets, min_support=0.03, min_confidence=0.3)


for rule in association_rules:
    items = list(rule.items)
    print(f"{items[:-1]}  --> [{items[-1]}]     Support: {rule.support:g}"
          + f"  Confidence: {rule.ordered_statistics[0].confidence:g}")    
```
